```python
sns.countplot(data = df, x='term',color='Red');
plt.title('Loan Term Distribution');
ax = sns.countplot(x='term', data=df,color='Red')
ax.bar_label(ax.containers[0], fmt='%d',fontsize=9,fontweight='light');
```

```python
sns.countplot(data = df, x='loanstatus',color='Red');
plt.title('Loan Status Distribution');
fig = plt.gcf()
plt.xticks(rotation=60);
fig.set_size_inches(10, 5)
ax = sns.countplot(x='loanstatus', data=df,color='Red')
ax.bar_label(ax.containers[0], fmt='%d',fontsize=9,fontweight='light');
```

```python
sns.countplot(data = df, x='creditgrade',color='Red');
plt.title('Credit Grade Distribution');
fig = plt.gcf()
fig.set_size_inches(10, 5)
```

```python
sns.countplot(data = df, x='occupation',color='Red');
plt.title('Occupation Distribution');
fig = plt.gcf()
plt.xticks(rotation=90);
fig.set_size_inches(12, 6)
```

```python
sns.countplot(data=df, x='employmentstatus', color='Red');
plt.title('Employment Status');
fig = plt.gcf()
fig.set_size_inches(10, 5)
ax = sns.countplot(x='employmentstatus', data=df,color='Red')
ax.bar_label(ax.containers[0], fmt='%d',fontsize=9,fontweight='light');
```

```python
sns.displot(df.loanoriginalamount, color='red');
plt.title('Loan Original Amount Distribution')
fig = plt.gcf()
fig.set_size_inches(10, 6)
```

## Bivariate Variables

In this analysis, the follwoing bivariate variables would be analyzed to derieve insights in the performance of prosper loan.

1. Recommendation by count of investment from friends: This would show the relationship between recommendation and investment made.

2. Loan status by interest rate: This will show the distribution of the 5 number summary; the minimum, first quartile, second quartile, median, third quartile and the maximum of the interest rates on the loan status.

3. Borrower APR by employemnt status: This will provide insight into the annual percentage rate paid by the categories of employment.

```python
df.info()
```

```python
sns.regplot(data=df, x='recommendations', y='investmentfromfriendscount', color= 'Red')
plt.title('Recommendation by Investment from friends')
fig = plt.gcf()
fig.set_size_inches(10, 6)
```

```python
sns.boxplot(data=df, x='loanstatus', y='borrowerrate', color= 'Red')
plt.xlabel('Loan Status')
plt.ylabel('Interest Rate')
plt.title('Loan Status by Interest Rate')
plt.xticks(rotation=60)
fig.set_size_inches(12, 6)
plt.show()
```

```python
sns.boxplot(data=df,y='borrowerapr',x='employmentstatus', color = 'Red');
plt.title('BorrowerAPR by Employment Status');
plt.ylabel('Borrower APR',size=10);
plt.xlabel('Employment Status',size=10);
fig.set_size_inches(10, 6)
plt.xticks(rotation=60);
```

```python
sns.regplot(data=df,x='monthlyloanpayment', y='lp_interestandfees', color = 'Red');
plt.title('Monthly loan payment by LP_interest and fees');
```
